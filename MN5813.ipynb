### Data management

#### Finding null values

### Cleaning column names

```python
sales.columns = (
    sales.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False))
```

All the columns of the dataset were cleaned by removing whitespace, converting to lowercase, and replacing special characters for error-free and consistent manipulation.  

### Missing value treatment

```python
categorical_cols = ['gender', 'color', 'sleeve_type',
    'material', 'combo_item']
sales[categorical_cols] = sales[categorical_cols].fillna('Unknown')
```

There are several categorical columns containing missing values, and the missing values are replaced with “unknown” to avoid loss of informative data from the sales data. 

### Converting discount(%) to numeric

```python
sales['discount'] = (
    sales['discount%']
    .str.replace('%', '', regex=False)
    .astype(float))
sales.drop(columns=['discount%'], inplace=True)
```

```python
# Replacing missing discounts with 0
sales['discount'] = sales['discount'].fillna(0)
# dropping missing value product_name
sales = sales.dropna(subset=['product_name'])
```

In the case of the “discount%” column, the field was cleaned by removing the percentage symbol and converting the column to a numeric format. The null values of the column were replaced with 0, and the value indicates that no discount is applied to the product sales. 

### Checking duplicate values

```python
sales.duplicated().sum()
```

### Removing duplicate values

```python
sales = sales.drop_duplicates()
```

It has been seen that in the dataset, there are 1641 rows containing duplicate data based on the product sales. Duplicate records were dropped to prevent data bias issues in the collected sales data.

### Data wrangling

```python
# price bands
sales['price_band'] = pd.qcut(
    sales['price'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Premium'])
```

```python
# Discount band
sales['discount_band'] = pd.qcut(
    sales['discount'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
# Sales per review
sales['sales_per_review'] = (
    sales['no._of_sells'] / (sales['no._of_review'] + 1))
```

The data wrangling method is used to create categorical columns using quartile segmentation, and this section allows for making comparisons of product performance based on pricing levels.

The data analysis highlights the mean product sales across different gender categories. The highest average sales were noticed in the female-targeted products, indicating that consumer demand isinfluenced by gender-specific marketing. Based on the dataset, relatively balanced market engagement was noticed across female and male segments. 

```python
sales.groupby('price_band')['no._of_sells'].mean()
```

It has been observed that high-priced products had thehighest average product sales, and the outcomes also suggest that medium and premium price category items have strong consumer demand. In this case, retailers can provide actionable insight based on pricing strategies, and value-driven pricing in premium and high segments can enhance the revenue of the business.

The outcomes demonstrate that a high discount band helped in producing the highest average sales, as well as low and medium discounts on the products decrease the product sales. Thus, it can be said that large discounts on products enhance product demand and influence consumer purchasing behaviour.
```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Average sales by price band
price_trends = sales.groupby("price_band")["no._of_sells"].mean()

# Discount band vs average sales
discount_trends = sales.groupby("discount_band")["no._of_sells"].mean()

# Relationship: reviews vs sales 
review_sales = sales[["no._of_review", "no._of_sells"]]

# Top 10 locations by average sales
top_locations = (
    sales.groupby("location")["no._of_sells"]
    .mean()
    .sort_values(ascending=False)
    .head(10))

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("Appendix – Summary of E-commerce Sales Patterns (Daraz)", 
             fontsize=16, fontweight="bold")

# Price Band vs Sales
price_trends.plot(kind="bar", ax=axes[0, 0], color="skyblue")
axes[0, 0].set_title("Average Sales by Price Band")
axes[0, 0].set_xlabel("Price Band")
axes[0, 0].set_ylabel("Average Number of Sales")
axes[0, 0].grid(True, alpha=0.3)

# Discount Band vs Sales
discount_trends.plot(kind="bar", ax=axes[0, 1], color="lightgreen")
axes[0, 1].set_title("Average Sales by Discount Band")
axes[0, 1].set_xlabel("Discount Band")
axes[0, 1].set_ylabel("Average Number of Sales")
axes[0, 1].grid(True, alpha=0.3)

# Reviews vs Sales Scatter
sns.scatterplot(
    x="no._of_review",
    y="no._of_sells",
    data=review_sales,
    alpha=0.5,
    ax=axes[1, 0]
)
axes[1, 0].set_title("Customer Reviews vs Number of Sales")
axes[1, 0].set_xlabel("Number of Reviews")
axes[1, 0].set_ylabel("Number of Sales")

# Top 10 Locations
top_locations.plot(kind="bar", ax=axes[1, 1], color="salmon")
axes[1, 1].set_title("Top 10 Locations by Average Sales")
axes[1, 1].set_xlabel("Location")
axes[1, 1].set_ylabel("Average Number of Sales")

plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.savefig("appendix_sales_dashboard.png", dpi=300, bbox_inches="tight")
plt.show()
```
